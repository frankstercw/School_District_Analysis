```python
import pandas as pd
import numpy as np

school_data_to_load = "Resources/schools_complete v2.csv"
student_data_to_load = "Resources/students_complete v2.csv"

school_data_df = pd.read_csv(school_data_to_load)
school_data_df

student_data_df = pd.read_csv(student_data_to_load)
student_data_df
```

```python
student_data_df.head()
```

```python
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]

for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"")


#update math and reading scores to NaN

student_data_df.loc[(student_data_df['grade'] == '9th') & (student_data_df['school_name'] == 'Thomas High School'), 'reading_score'] = np.nan
student_data_df.loc[(student_data_df['grade'] == '9th') & (student_data_df['school_name'] == 'Thomas High School'), 'math_score'] = np.nan

student_data_df.tail(10)

```

```python
school_student_data_df = pd.merge(student_data_df, school_data_df, on=["school_name", "school_name"])
school_student_data_df.head()

student_count = school_student_data_df["Student ID"].count()
# student_count

school_count_unique = school_student_data_df["school_name"].unique()
school_count = len(school_count_unique)
school_count

total_budget = school_data_df["budget"].sum()
# total_budget

average_reading_score = school_student_data_df["reading_score"].mean()
# average_reading_score

average_math_score = school_student_data_df["math_score"].mean()
# average_math_score

passing_math = school_student_data_df[school_student_data_df["math_score"] >= 70]
passing_reading = school_student_data_df[school_student_data_df["reading_score"] >= 70]


passing_math_count = passing_math["student_name"].count()
passing_reading_count = passing_reading["student_name"].count()

# passing_math_count

passing_math_percentage = passing_math_count / float(student_count) * 100

# passing_math_percentage

passing_reading_percentage = passing_reading_count / float(student_count) * 100

passing_math_reading = school_student_data_df[(school_student_data_df["math_score"] >= 70) & (school_student_data_df["reading_score"] >= 70)]

overall_passing_math_reading_count = passing_math_reading["student_name"].count()
# overall_passing_math_reading_count

overall_passing_percentage = overall_passing_math_reading_count / student_count * 100
# overall_passing_percentage

district_summary_df = pd.DataFrame(
          [{"Total Schools": school_count,
          "Total Students": student_count,
          "Total Budget": total_budget,
          "Average Math Score": average_math_score,
          "Average Reading Score": average_reading_score,
          "% Passing Math": passing_math_percentage,
         "% Passing Reading": passing_reading_percentage,
        "% Overall Passing": overall_passing_percentage}])
district_summary_df
```
